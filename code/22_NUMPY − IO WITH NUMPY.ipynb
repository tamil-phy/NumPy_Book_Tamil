### 22. NUMPY − I/O WITH NUMPY

NumPy-யில் data-ஐ save மற்றும் load செய்ய பல methods உள்ளன, இதனால் data-ஐ file
format-ல் எளிதாக சேமிக்கவும், retrieve செய்யவும் முடியும். Data I/O operations
scientific computing மற்றும் data analysis-ல் மிகவும் முக்கியமானவை.

#### 22.1. numpy.save()

**save()** function-ஐ பயன்படுத்தி, array-ஐ binary format-ல் (.npy)
சேமிக்கலாம். இது data-ஐ compact-ஆகவும் memory-efficient-ஆகவும் சேமிக்க
உதவுகிறது.

**உதாரணம்:**

``` python
import numpy as np

# Saving array to file
array = np.array([1, 2, 3, 4, 5])
np.save('my_array', array)  # Save the array as a .npy file
```

**Output:**

    Array saved as 'my_array.npy' in the current directory.

#### விளக்கம்:

-   **np.save()** function-ஐ array-ஐ binary format-ல் சேமிக்க
    பயன்படுத்துகிறோம்.
-   இந்த method data-ஐ efficient-ஆக save செய்து, future usage-க்கு compact
    format-ல் வைத்திருக்கிறது.

#### 22.2. numpy.savetxt()

**savetxt()** function-ஐ பயன்படுத்தி, array-ஐ text file format-ல்
சேமிக்கலாம். Text format-ல் data-ஐ human-readable-ஆக save செய்ய இது பயன்படும்.

**உதாரணம்:**

``` python
# Saving array to text file
np.savetxt('my_array.txt', array, delimiter=',')  # Save the array as a .txt file
```

**Output:**

    Array saved as 'my_array.txt' with comma-separated values.

#### விளக்கம்:

-   **np.savetxt()** function array-ஐ text file format-ல் (.txt)
    சேமிக்கிறது.
-   Text file format data-ஐ human-readable-ஆக save செய்ய உதவுகிறது, இது
    data analysis மற்றும் data sharing-க்கு பயனுள்ளதாக இருக்கும்.
